```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# URL of page to be scraped
url = 'https://washingtondc.craigslist.org/search/sss?sort=rel&query=lysol'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# results are returned as an iterable list
results = soup.find_all('li', class_="result-row")
```

```python
print(results[0].prettify())
```

```python
results_list = []

# Loop through returned results
for result in results:
    
    title = ''
    price = ''
    link = ''
    location = ''
    
    # Error handling
    try:
        # Identify and return title of listing
        title = result.find('a', class_="result-title").text
        # Identify and return price of listing
        price = result.find('span', class_="result-price").text
        # Identify and return link to listing
        link = result.find('a', class_="result-title")['href']
        # Identify and location
        location = result.find('span', class_="result-hood").text
            
    except Exception as e:
        print(e)
        
    result_dict = {
        'Title': title,
        'Price': price,
        'Link': link,
        'Location': location
    }
        
    results_list.append(result_dict)
```

```python
pd.set_option('display.max_rows', None)
```

```python
items_df = pd.DataFrame(results_list)
items_df['Location'] = items_df['Location'].str.strip().str.lstrip('(')
items_df['Location'] = items_df['Location'].str.strip().str.rstrip(')')
items_df['Price'] = items_df['Price'].str.strip('$').astype(float)
items_df
```
